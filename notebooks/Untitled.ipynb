```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_y = ((torch.rand(60)**3) - 0.5) * 2
raw_y, _ = torch.sort(raw_y)
plt.plot(raw_y)
plt.show()


y = torch.softmax(raw_y, dim=-1)

plt.plot(y)
plt.show()

y = torch.softmax(raw_y * 2, dim=-1)

plt.plot(y)
plt.show()

y = torch.softmax(raw_y * 4, dim=-1)

plt.plot(y)
plt.show()

y = torch.softmax(raw_y * 8, dim=-1)

plt.plot(y)
plt.show()
```

```python
def trial(size, exp, rm=1.0):
    raw_y = ((torch.rand(size)**exp) - 0.5) * 2
    raw_y, _ = torch.sort(raw_y)
    ran = (torch.rand(60) - 0.5) * 2
    ran = ran * rm
    y = raw_y + ran
    m = torch.argmax(y)
    return m, raw_y[-1] - raw_y[m]

t = 0.0
for i in range(100):
    _, delta = trial(60, 3, rm=1.0)
    t += delta
print(t / 100)

for i in range(1000):
    rm = 10.0 / (i + 1)
    t = 0
    for _ in range(10000):
        _, delta = trial(60, 3, rm=rm)
        t += delta
    t = t / 10000.0
    print(rm, t)
```
