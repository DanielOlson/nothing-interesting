```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


def sxysort_ground(x, y):
    vals = np.add.outer(x, y)
    vals = vals.reshape(-1)
    vals = np.sort(vals)
    return vals
    
size = 10000

a = np.sort(np.random.rand(size))
b = np.sort(np.random.rand(size))

y = sxysort_ground(a, b)
x = np.arange(len(y))

xs = np.arange(len(a))

plt.plot(xs, a)
plt.plot(xs, b)
plt.show()
plt.plot(x, y)
plt.show()


```

```python
fa = np.fft.fft(a)
fb = np.fft.fft(b)
fy = np.fft.fft(y)


plt.plot(fa.real)
plt.plot(fa.imag)
plt.show()

plt.plot(fb.real)
plt.plot(fb.imag)
plt.show()

plt.plot(fy.real)
plt.plot(fy.imag)
plt.show()
```

```python
size = 10

a = np.sort(np.random.rand(size))
b = np.sort(np.random.rand(size))

oaa = np.add.outer(a, a)
obb = np.add.outer(b, b)
oy = np.add.outer(a, b)


plt.imshow(oaa)
plt.show()

plt.imshow(obb)
plt.show()

plt.imshow(oy)
plt.show()
```

```python
foy = np.fft.fft2(oy)
print(foy.shape)
plt.imshow(foy.real)
plt.show()
plt.imshow(foy.imag)
plt.show()
```
